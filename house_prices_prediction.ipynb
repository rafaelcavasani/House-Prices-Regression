# Random Forest

O Random Forest é um modelo de machine learning baseado em arvores de decisão. Ele gera várias árvores de decisão com base nos dados de treinamento para predizer os valores na base de validação.

## Importando as Bibliotecas

Irei utilizar o sklearn para a criação do modelo e o sckiti optimizer para encontrar os melhores parâmetros para o modelo.

Para isso, é necessário instalar o sckiti optimizer (skopt)

```python
!pip install scikit-optimize
```

```python
from skopt import gp_minimize
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

O código abaixo divide o atributo que queremos prever (SalesPrice) dos demais atributos. Por convenção, chamei nossos previsores de X e nossa classe de y. 

```python
X = df_final.iloc[:, 1:67].values
y = df_final.iloc[:, 67].values
```

Alguns modelos de machine learning performam melhor com os dados em uma escala padrão. Então vou criar um objeto StandardScale do sklearn para escalonar os dados e testar qual maneira o algoritmo se sai melhor.

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

O método train_test_split do sklearn divide o dataset em duas partes para treinarmos e validarmos o nosso modelo. Definimos o tamanho da divisão no parâmetro test_size, no caso, 30% será para a validação e 70% para o treinamento.

Esse método retorna quatro valores:
* X_train: 70% dos previsores para o treinamento
* X_test: 30% dos previsores para a validação
* y_train: 70% das classes, relacionadas a parte de treinamento
* y_test: 30% das classes, relacionadas a validação

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Agora, vamos criar uma função que define a criação do modelo RandomForestRegressor, o treinamento com nossos atributos previsores X_train e a validação dos dados com os atributos previsores X_test.

Passamos a variável **params** como parâmetro para a função.

A função de avaliação do erro do modelo escolhido foi o **mean_squared_error**. Essa função retorna a diferença quadrada média entre o valor real e o valor que foi predito pelo modelo.

```python
def train_model(params):
    estimators = params[0]
    max_depth = params[1]
    min_samples_split = params[2]
    min_samples_leaf = params[3]
    max_features = params[4]
    min_impurity_decrease = params[5]
    max_samples = params[6]
    ccp_alpha = params[7]
    print('\n')
    random_forest = RandomForestRegressor(n_estimators=estimators,
                                          criterion='mse', 
                                          max_depth=max_depth, 
                                          min_samples_split=min_samples_split,
                                          min_samples_leaf=min_samples_leaf,
                                          max_features=max_features,
                                          min_impurity_decrease=min_impurity_decrease,
                                          max_samples=max_samples, ccp_alpha=ccp_alpha)
    random_forest.fit(X_train, y_train)
    predictions = random_forest.predict(X_test)

    return mean_squared_error(y_test, predictions)
```

Criamos o parâmetro **params** para atriburi a função gp_minimize e informamos o range de cada parâmetro que queremos ajustar no modelo. O gp_minimize irá pegar um valor aleatório entre o range passado a cada chamada e validar o modelo, assim, ao fim da execução, teremos o melhor resultado e quais foram os parâmetros utilizados para esse resultado.

Passamos os seguintes parâmentros para a função gp_minimize:
* func: A função que irá criar e treinar o modelo a cada chamada
* dimensions: A lista com os ranges de cada parâmetro
* random_state: Para plantarmos a semente aleatória e termos sempre o mesmo resultado
* verbose: Para que imprima tudo o que está sendo feito
* n_calls: Número de chamadas ou vezes que o modelo será criado e validado
* n_random_state: Número de vezes que os parâmetros serão inicializados a cada chamada da função

Após a execução, podemos verificar quais foram os parâmetros que deram os melhores resultado com o atributo x do retorno do gp_minimize.

### Parâmetros Random Forest:

* **estimators**: Número de árvores de decisão na Random Forest
* **max_deep**: Profundidade máxima de cada árvore
* **min_samples_split**: O número mínimo de amostras necessárias para dividir um nó interno
* **min_samples_leaf**: O número mínimo de amostras necessárias para estar em um nó
* **max_features**: O número de recursos a serem considerados ao procurar a melhor divisão

Função para criar a rede neural.

Criaremos uma rede neural simples com o tensorflow. Para isso, instanciamos o modelo sequencial do keras e adicionamos as camadas abaixo. Adicionamos uma camada densa com 24 neurônios de entrada, 12 neurônios na camada interna e função de ativação relu.

Logo após, adicionamos uma camada de Dropout. Essa camada inativa alguns neurônios aleatoreamente para evitar overfiting.

Por fim, adicionamos a camada de saída com 1 neurônio e função de ativação linear, compilamos e retornamos nosso modelo compilado.

```python
def nn_create():
    regressor = tf.keras.models.Sequential()
    regressor.add(tf.keras.layers.Dense(units=12, activation='relu', input_dim=24))
    regressor.add(tf.keras.layers.Dropout(0.1))
    regressor.add(tf.keras.layers.Dense(units=1, activation='linear'))
    regressor.compile(optimizer='adam', loss='mse', metrics=['mse'])
    return regressor
```

Instanciando o modelo na variavel nn_regressor

```python
nn_regressor = tf.keras.wrappers.scikit_learn.KerasRegressor(build_fn = nn_create,
                                                           epochs = 1000,
                                                           batch_size = 30)
```

Treinamento da rede neural

```python
nn_regressor.fit(X_features, y)
```
